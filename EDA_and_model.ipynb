```python
# Distance between pickup and dropoff

def haversine_np(lon1, lat1, lon2, lat2):
    lon1, lat1, lon2, lat2 = map(np.radians, [lon1, lat1, lon2, lat2])
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    a = np.sin(dlat/2.0)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon/2.0)**2
    c = 2 * np.arcsin(np.sqrt(a))
    km = 6367 * c
    return km

df['distance_km'] = haversine_np(df['pickup_longitude'], df['pickup_latitude'], 
                                 df['dropoff_longitude'], df['dropoff_latitude'])
```

# Model Preprocessing

```python
from sklearn.preprocessing import OneHotEncoder , StandardScaler
from sklearn.model_selection import train_test_split
import pickle
```

```python
df['hour'].unique()
```

```python
#1. Make ALL posible categoies list for 5 columns
hours_list = list(range(24))                                            # 0 se 23
days_list = list(range(7))                                              # 0 se 6
months_list = list(range(1, 13))                                        # 1 se 12
passengers_list = [1 , 2 , 3 , 4 , 5 , 6]                               # 1 se 6
time_slots_list = ["Night", "Morning", "Afternoon", "Evening"] 

#2. Make list of list for all this lists
all_categories = [hours_list, days_list, months_list, passengers_list, time_slots_list]

# 3. OneHotEncoder initialize 
encoder = OneHotEncoder(categories=all_categories, sparse_output=False, handle_unknown='error')

# 4. Make the list of my columns name
categorical_cols = ['hour', 'day_of_week', 'month', 'passenger_count', 'time_slots']

# 5. Apply fit_transform
encoded_array = encoder.fit_transform(df[categorical_cols])

# 6. Make dataframe for this new columns
encoded_df = pd.DataFrame(encoded_array, columns=encoder.get_feature_names_out(categorical_cols))

# 7 . Drop the original 5 columns and concat this new dataframe with our original old dataframe
df_remaining = df.drop(categorical_cols, axis=1).reset_index(drop=True)
df_final = pd.concat([df_remaining, encoded_df], axis=1)
```

```python
X = df_final
```

```python
X_train , X_test , y_train , y_test = train_test_split(
    X , y , test_size=0.25 , random_state=42
)
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

# ANN regression Model trianing

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential([
    Dense(64 , activation='relu' , input_shape=(X_train.shape[1],)),
    Dense(32 , activation='relu'),
    Dense(1)            #Output layer for regression don't need activation function
])
```

```python
def create_model(neurons=32 , layers=1):
    model = Sequential()
    model.add(Dense(neurons , activation='relu' , input_shape=(X_train.shape[1],)))

    for _ in range(layers-1):
        model.add(Dense(neurons , activation='relu'))

    model.add(Dense(1))         #Output layer for regression don't need activation function

    # Compile the model
    model.compile(optimizer='adam' , loss='mean_absolute_error' , metrics=['mae'])

    return model

```

```python
# Create Keras Regressor Model
from scikeras.wrappers import KerasRegressor

model = KerasRegressor(layers =1 , neurons=32,model=create_model , verbose=1)
```

```python
# Define the Grid search paramenter
param_grid = {
    'neurons' : [16 , 32 , 64 , 128],
    'layers' : [1,2],
    'epochs' : [50 ,100]
}
```

```python
# Perfornm Grid Search
from sklearn.model_selection import GridSearchCV

grid = GridSearchCV(estimator=model , param_grid=param_grid , n_jobs=-1 ,verbose=1, cv=3)
grid_result = grid.fit(X_train , y_train)
```
